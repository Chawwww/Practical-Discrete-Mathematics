```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# find the shape of the auto dataset
df = pd.read_csv("diploma_students_height_weight.csv")
df.shape
```

```python
Y = df["Weight"]
X = df["Height"]  # Changed 'heigth' to 'Height'

X_b = np.c_[np.ones((100,1)),X] #here we are adding X_o = 1 to all the feature values
beta_values = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(Y)

print(beta_values)
```

```python
X_new = np.array([[158]])
# Changed the shape of np.ones to (1, 1) to match X_new
X_new_b = np.c_[np.ones((1, 1)), X_new]
y_predict = X_new_b.dot(beta_values)

# Changed index to 0 to access a valid element of y_predict
print(f"height of a person =158 cm; predicted weight is {y_predict[0]:.3f}; ")
```

```python
#x figure
X_plot= np.array([[158]])
X_plot_b = np.c_[np.ones((1,1)),X_plot]
Y_plot = X_plot_b.dot(beta_values)

Equationline = "Y = {:.3f}+{:.3f}X".format(beta_values[0], beta_values[1])

plt.plot(X_plot, Y_plot, "r-", label = Equationline)

# Create a DataFrame from X and Y for scatterplot
data = pd.DataFrame({'X': X, 'Y': Y})  # Assuming X and Y are pandas Series
sns.scatterplot(x='X', y='Y', data=data, label="Training Data")  # Pass data and specify x, y columns

plt.legend()
plt.show()
```
